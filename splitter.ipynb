Write a Python program that reads a string where multiple numbers are separated by a separator and reads a string that specifies the separator.

Now convert the String into a list **without using the split** function. The type of list values should be an integer.

Sample input 1:

    7,5,6,8,2,

    ,

Sample output 1:

    [7,5,6,8,2]
    

```python
n=input()
sp=input()

l=[]
s=0
for i in range(len(n)):
  if n[i]==sp:
    f=i
    x=n[s:f]
    l.append(int(x))
    s=i+1
print(l)
```
