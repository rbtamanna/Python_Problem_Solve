Write a Python program that takes a String as an input from the user and counts the frequency of each character using a dictionary. [You are not allowed to use the count() function]
===================================================================
Sample Input:
"Python programming is fun"
Sample Output:
{'p': 2, 'y': 1, 't': 1, 'h': 1, 'o': 2, 'n': 3, 'r': 2, 'g': 2, 'a': 1, 'm': 2, 'i': 2, 's': 1, 'f': 1, 'u': 1}
===================================================================


```python
s=input()
d={}
s=s.lower()
for i in range(len(s)):
  if s[i]!=' ':
    if s[i] in d:
      d[s[i]]+=1
    else :
      d[s[i]]=1
print(d)
```
